```python
%matplotlib inline
from argparse import Namespace
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
#from dask import delayed
#from dask.distributed import Client

torch.set_printoptions(edgeitems=2, linewidth=75)
```

```python
t_c = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c).unsqueeze(1) # <1>
t_u = torch.tensor(t_u).unsqueeze(1) # <1>

t_u.shape
```

```python
n_samples = t_u.shape[0]
n_val = int(0.2 * n_samples)

shuffled_indices = torch.randperm(n_samples)

train_indices = shuffled_indices[:-n_val]
val_indices = shuffled_indices[-n_val:]

train_indices, val_indices
```

```python
t_u_train = t_u[train_indices]
t_c_train = t_c[train_indices]

t_u_val = t_u[val_indices]
t_c_val = t_c[val_indices]

t_un_train = 0.1 * t_u_train
t_un_val = 0.1 * t_u_val
```

```python
def training_loop(n_epochs, optimizer, model, loss_fn, t_u_train, t_u_val,
                 t_c_train, t_c_val):
    for epoch in range(1, n_epochs+1):
        t_p_train = model(t_u_train)
        loss_train = loss_fn(t_p_train, t_c_train)
        
        t_p_val = model(t_u_val)
        loss_val = loss_fn(t_p_val, t_c_val)
        
        optimizer.zero_grad()
        loss_train.backward()
        optimizer.step()
        
        if (epoch == 1) or (epoch % 1000 == 0):
            print(f'Epoch {epoch}, Training loss {loss_train.item():.4f},'
                 f' Validation loss {loss_val.item():.4f}')
```

```python
linear_model = nn.Linear(1, 1)
optimizer = optim.SGD(linear_model.parameters(), lr=1e-2)

training_params = Namespace(
    n_epochs=3000,
    optimizer=optimizer,
    model=linear_model,
    loss_fn=nn.MSELoss(),
    t_u_train=t_un_train,
    t_u_val=t_un_val,
    t_c_train=t_c_train,
    t_c_val=t_c_val)

training_loop(
    n_epochs=training_params.n_epochs,
    optimizer=training_params.optimizer,
    model=training_params.model,
    loss_fn=training_params.loss_fn,
    t_u_train=training_params.t_u_train,
    t_u_val=training_params.t_u_val,
    t_c_train=training_params.t_c_train,
    t_c_val=training_params.t_c_val)

print()
print(linear_model.weight)
print(linear_model.bias)
```

```python
seq_model = nn.Sequential(
    nn.Linear(1, 13),
    nn.Tanh(),
    nn.Linear(13, 1))

seq_model
```

```python
[param.shape for param in seq_model.parameters()]
```

```python
for name, param in seq_model.named_parameters():
    print(name, param.shape)
```

```python
from collections import OrderedDict

seq_model = nn.Sequential(OrderedDict([
    ('hidden_linear', nn.Linear(1, 8)),
    ('hidden_activation', nn.Tanh()),
    ('output_linear', nn.Linear(8, 1))
    ]))

seq_model
```

```python
for name, param in seq_model.named_parameters():
    print(name, param.shape)
```

```python
seq_model.output_linear.bias
```

```python
optimizer = optim.SGD(seq_model.parameters(), lr=1e-3)

training_params = Namespace(
    n_epochs=5000,
    optimizer=optimizer,
    model=seq_model,
    loss_fn=nn.MSELoss(),
    t_u_train=t_un_train,
    t_u_val=t_un_val,
    t_c_train=t_c_train,
    t_c_val=t_c_val)

training_loop(
    n_epochs=training_params.n_epochs,
    optimizer=training_params.optimizer,
    model=training_params.model,
    loss_fn=training_params.loss_fn,
    t_u_train=training_params.t_u_train,
    t_u_val=training_params.t_u_val,
    t_c_train=training_params.t_c_train,
    t_c_val=training_params.t_c_val)

print('output', seq_model(t_un_val))
print('answer', t_c_val)
print('hidden', seq_model.hidden_linear.weight.grad)
```

```python
#comparing to linear model
from matplotlib import pyplot as plt

t_range = torch.arange(20., 90.).unsqueeze(1)

fig = plt.figure(dpi=600)
plt.xlabel("Fahrenheit")
plt.ylabel("Celsius")
plt.plot(t_u.numpy(), t_c.numpy(), 'o')
plt.plot(t_range.numpy(), seq_model(0.1 * t_range).detach().numpy())
plt.plot(t_u.numpy(), seq_model(0.1 * t_u).detach().numpy(), 'kx')
```

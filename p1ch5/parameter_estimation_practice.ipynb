```python
%matplotlib inline
import numpy as np
import torch
torch.set_printoptions(edgeitems=2, linewidth=75)
```

```python
t_c = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
```

```python
def model(t_u, w, b):
    '''Takes the input parameter, weight parameter, and bias parameter and
    returns the result.
    
    params:
    
    t_u: tensor in unknown units
    '''
    return w * t_u + b 
```

```python
def loss_fn(t_p, t_c):
    '''A value that we want to minimize, in this case the squared difference
    of the two numbers. We produce a tensor with all of the differences, then
    square those values, and then obtain the average. This is known as the mean
    squared loss.'''
    squared_diffs = (t_p - t_c)**2
    return squared_diffs.mean()
```

```python
delta = 0.1
loss_rate_of_change_w = \
    (loss_fn(model(t_u, w + delta, b), t_c) - 
    loss_fn(model(t_u, w - delta, b), t_c)) / (2.0 * delta)
```

```python
learning_rate = 1e-2
w = w - learning_rate * loss_rate_of_change_w
```

```python
def dloss_fn(t_p: torch.Tensor, t_c: torch.Tensor):
    '''Takes the predicted temperatures and the celsius
    temperatures and returns the derivative'''
    dsq_diffs = 2 * (t_p - t_c) / t_p.size(0)
    return dsq_diffs
```

```python
def dmodel_dw(t_u, w, b):
    return t_u
```

```python
def dmodel_db(t_u, w, b):
    return 1.0
```

```python
def grad_fn(t_u: torch.Tensor, t_c: torch.Tensor,
            t_p: torch.Tensor, w: torch.Tensor,
            b: torch.Tensor):
    '''Applies the gradient function to the given inputs.'''
    dloss_dtp = dloss_fn(t_p, t_c)
    dloss_dw = dloss_dtp * dmodel_dw(t_u, w, b)
    dloss_db = dloss_dtp * dmodel_db(t_u, w, b)
    return torch.stack([dloss_dw.sum(), dloss_db.sum()])
```

```python
def training_loop(n_epochs: int, learning_rate, params, t_u, t_c):
    '''The training loop runs the data through a number of epochs to train it.'''
    for epoch in range(1, n_epochs+1):
        w, b = params
        t_p = model(t_u, w, b) #forward pass
        loss = loss_fn(t_p, t_c)
        grad = grad_fn(t_u, t_c, t_p, w, b) #backward pass
        params = params - learning_rate * grad
        print('Epoch %d, Loss %f' % (epoch, float(loss)))
    return params
```

```python
training_loop(
    n_epochs=100,
    learning_rate = 1e-2,
    params = torch.tensor([1.0, 0.0]),
    t_u = t_u,
    t_c = t_c
)
```

```python
training_loop(
    n_epochs=100,
    learning_rate = 1e-4,
    params = torch.tensor([1.0, 0.0]),
    t_u = t_u,
    t_c = t_c
)
```

```python
#normalize t_u to ensure accurate calculations
t_un = 0.1 * t_u
```
